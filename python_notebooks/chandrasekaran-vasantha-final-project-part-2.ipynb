## PLOT-1: Driver Plot
### Plot showing the count of free and paid games according to their month of release


```python
chart_1 = alt.Chart.from_dict({
    "data":{"url":"https://raw.githubusercontent.com/Saahil87/Saahil87.github.io/main/_data/games-features.csv"},
  "mark":"bar",
  "height":500,
  "width":500,
  "encoding":{
    "x":{
      "timeUnit":"month",
      "field": "ReleaseDate",
      "type": "ordinal",
      "title": "Release Month"
    },
    "y":{
        "aggregate":"count",
        "type":"quantitative"
    },
    "color": {"field": "IsFree","type": "ordinal"}
          
  }
            
          })

chart_1
```

```python
chart_1.properties(width='container').save(myJekyllSaveDir+'project-chart1.vl.json')
```

### DRIVER PLOT :
Chart_1, shown above is the "Driver" plot.
#### Description:
Chart_1  is a bar chart that shows the count of records of 'free' and 'paid' games which are of the domain 'True' and 'False' of the 'IsFree' column, with respect to the month of release of that game.

The x-axis includes the release months of the game from January to December. There are few values whose release month is not specified,and they correspond to the unnamed value.

The y-axis includes the count of records of type -'IsFree == True' (dark blue) and 'IsFree== False'(light blue).
The count of records of free and paid games can be calculated and can also be compared according to their release months.
#### Aesthetics:
Labels: The x and y labels are properly defined as 'Release Month' and 'Count of Records' respectively.

Color Scheme: The color of the values is chosen to be default- Blue. The dark and light blue colors represent the count of records of free and paid games respectively.

Scale: The scales representing the free(IsFree-TRUE) and paid games (IsFree-FALSE) and its color scheme is represented at the top-right corner of the driver plot.

## Dashboard:
The above two charts are linked to create a dashboard. Interactivity is achieved using the 'brush' tool and the changes made due to the selection of a particular interval in chart-1 are reflected in chart-2.

```python
brush = alt.selection_interval(encodings=['x','y'])
chart_1 = alt.Chart.from_dict({
    "data":{"url":"https://raw.githubusercontent.com/Saahil87/Saahil87.github.io/main/_data/games-features.csv"},
  "mark":"bar",
  "height":500,
  "width":500,
  "encoding":{
    "x":{
      "timeUnit":"month",
      "field": "ReleaseDate",
      "type": "ordinal",
      "title": "Release Month"
    },
    "y":{
        "aggregate":"count",
        "type":"quantitative"
    },
    "color": {"field": "IsFree","type": "ordinal"}
          
  }
            
          }).add_selection(
        brush
    )
chart_2 = alt.Chart.from_dict({
  "data":{"url":"https://raw.githubusercontent.com/Saahil87/Saahil87.github.io/main/_data/games-features.csv"},
  "mark":"line",
    "height":500,
    "width":500,
  "encoding":{
    "x":{"timeUnit":"month","field": "ReleaseDate","type": "ordinal","title": "Release Month"},
    "y":{"field":"AchievementCount","type":"quantitative","axis":{"title":" Maximum Achievement Count"},"aggregate":"max"}
  }}
).transform_filter(
        brush
    )
chart = chart_1 | chart_2
```

```python
chart
```

```python
chart.save(myJekyllSaveDir+'project-dashboard.vl.json')
```

## Dashboard description:
The dashboard shown above has two plots -each showing the number of Free, Paid Games in the Steam Game Platform and the Maximum number of Achievement in games with respect to the month of release of games.

The above chart is interactive. If the mouse pointer is clicked and dragged to select an interval of months in the left chart, it can be seen that the line chart in the right changes according to the interval selected.

The left chart shows the number of free and paid games in the Steam Game platform shown in light blue (Paid games) and dark blue(free games) bars. The number of free and paid games in each month of release of game can also be observed.

The right chart shows the maximum achievement count of games in the Steam Game platform as a line chart. The peaks represent the maximum achievement count in each release month of the games.

For instance, the left chart shows that during the release months- June and August, the number of free games were high. If the selection is made for months- June, July,August in the left chart, the corresponding maximum achievement count for those months can be seen from the line chart shown on the right. The right chart shows that the achievement count peaked during the month of June with 1150 within the selection interval from June to August.

Similar analysis can be made for the maximum achievement count for paid games, maximum achievement count for free games and comparing them with their month of release can be done using the dashboard and the trends according to the release month, achievement count, free/paid games can be visualized. This might be useful for the Steam Game Platform to improve their gaming service.

### Steam Video Game Hours Played Dataset
DATASET: An interesting contextual dataset that would add context to the Steam Game Dataset used would be ' Steam Video Game Hours Played' dataset.

CITATION: Steam Video Games. (2017, March 9). Kaggle. https://www.kaggle.com/datasets/tamber/steam-video-games

LICENSE: http://opendatacommons.org/licenses/dbcl/1.0/

### Description-Contextual Dataset:
The Steam Video Game Hours Played dataset includes a list of the user behaviors with columns- user ID, game title, the behavior including 'purchase' and 'play'.

The dataset used for the project was the Steam Game dataset which includes information about the games in the Steam Game platform with hardware requirements,  different genre of games, currency used, recommendation count, release date, response count,achievement count etc., but there is no information about the behavior of users in the platform with respect to the number of hours played by users.

The contextual dataset- 'Steam Video Game Hours Played' which includes information about the number of hours played by gamers in the Steam Game platform would be useful to visualize the overall user activity and the specification of the Steam Game platform along with the original dataset, Steam Game Dataset.
This would benefit the developers of the Steam Game platform to improve their overall interface and gamers would be able to see game statistics and individual statistics.
